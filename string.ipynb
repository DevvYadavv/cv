## STRING


```python
# a string is a sequence of characters enclosed in quotes 
s1 = "Hello"
s2 = 'World'
```

```python
# type check 
print(type(s1))  # <class 'str'>
```

```python
s = "DevYadav"
```

```python
print(s[0])   # D
print(s[3])   # Y
print(s[-1])  # v
print(s[-2])  # a
```

```python
# String slicing 
s = "DevYadav"

print(s[0:3])   # Dev
print(s[:3])    # Dev
print(s[3:])    # Yadav
print(s[::2])   # DvYdv
print(s[::-1])  # Reverse string
```

```python
#loop though string
s = "Dev"

for ch in s:
    print(ch)
```

```python
for i in range(len(s)):
    print(s[i])
```

```python
s = "DevYadav"
print(s.lower())  # devyadav
print(s.upper())  # DEVYADAV
```

```python
# removes space 
s = "  Dev  "
print(s.strip())
```

```python
#replace 
s = "Dev Yadav"
print(s.replace("Dev", "Mr"))
```

```python
# find 
s = "DevYadav"
print(s.find("Y"))   # 3
print(s.find("Z"))   # -1
```

```python
# if you want to find many y 
s = "DevYYadav"

for i in range(len(s)):
    if s[i] == "Y":
        print(i)
```
